```octave
Kb=10^-2.981846; BT=0.0145;
a=1;
b=Kb;
c=-Kb*BT;
t=roots([a b c]); //finds the roots
t=t(imag(t)==0); //sets any imaginary roots to zero
// display the positive real root

t=t(t>0)

// take minus log in case that might be useful (pH for example)

pt=-log10(t)
```

```octave
c.ScilabKernel.plot_settings = dict(format='svg')
x=[-10*t:t/100:t*10];
y=a*x.^2+b.*x+c;

plot(x,y,'b-','linewidth',2)
set(gca(),"auto_clear","off")
plot([t t],[min(y)-0.1*max(y) max(y)],'r','linewidth',2)
plot(x,zeros(size(x,1),size(x,2)),'k','linewidth',2)
plot([0 0],[min(y)-0.1*max(y) max(y)],'k','linewidth',2)

h=get("current_axes");//get the handle of the newly created axes
h.font_size=3; //set the tics label font size
h.thickness=2; //axis thickness
h.data_bounds=[0 min(y); max(x) max(y)];

xlabel('x values','fontsize',4); ylabel('y values','fontsize',4)
```
